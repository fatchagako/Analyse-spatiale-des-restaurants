**II. Extraction des Latitudes & longitudes**

```python
# installation de la bibliothéque geopy
!pip install geopy
```

```python
# service de géocodage utilisé pour convertir des adresses en coordonnées géographiques (latitude et longitude) et vice versa.
from geopy.geocoders import Nominatim
import random
```

```python
# Créer un objet Nominatim
# Créer un objet Nominatim avec un timeout augmenté
geolocator = Nominatim(user_agent="app", timeout=10)
```

```python
import numpy as np
import pandas as pd
from geopy.geocoders import Nominatim
import time

lat = [] 
lon = []


for name in locations['name']:
    location = geolocator.geocode(name)
    
    if location is None:
        lat.append(np.nan)
        lon.append(np.nan)
        
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)
```

```python
print(lat,lon)
```

```python
locations['lat'] = lat
locations['lon'] = lon
locations.head()
```

```python
locations.to_csv(r'C:\Users\User\Desktop\1..Zomato Spatial Analysis\zomato_locations.csv',index=False)
```

```python
# création de la variable location restau
Rest_locations = data['location'].value_counts().reset_index()
Rest_locations
```

```python
print(Rest_locations.columns)
print(locations.columns)
```

```python
# Effectuer la jointure en utilisant 'location' et 'name'
Restaurant_locations = Rest_locations.merge(locations, left_on='location', right_on='name', how='left').dropna()
Restaurant_locations
```

```python
data['rating'] = data['rate'].str.split('/', expand=True)[0]
```

```python
#identifions  toutes les notes ou évaluations
data['rating'].unique()
```

```python
# Remplacons les valeurs 'NEW'  et '-' par 0 
data.replace('NEW' , '0' , inplace=True)
data.replace('-' , '0' , inplace=True)
```

```python
data.dtypes
```

```python
data['rating']= pd.to_numeric(data['rating'])
```

```python
#Obtenons le classement des localisations en fonction de leur note moyenne dans le DataFrame 
data.groupby('location')['rating'].mean().sort_values(ascending=False)
```

```python
# Obtenons les moyennes de notes pour chaque localisation
avg_rating =data.groupby('location')['rating'].mean().sort_values(ascending=False).values
avg_rating
```

```python
# Obtenons la liste des localisations triées par ordre décroissant de leur note moyenne.
location = data.groupby('location')['rating'].mean().sort_values(ascending=False).index
location
```

```python
rating = pd.DataFrame()
```

```python
lat = [] 
lon = []


for loc in location:
    loc = geolocator.geocode(loc)
    
    if loc is None:
        lat.append(np.nan)
        lon.append(np.nan)
        
    else:
        lat.append(loc.latitude)
        lon.append(loc.longitude)
```

```python
rating['location']= location
rating['lat']= lat
rating['lon']= lon
rating['avg_rating']= avg_rating
rating.head()
```

```python
rating.isnull().sum()
```

```python
rating.dropna(inplace= True)
```

```python
HeatMap(rating[['lat', 'lon' , 'avg_rating']]).add_to(basemap)
basemap
```
